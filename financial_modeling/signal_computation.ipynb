Example: compute a bollinger band

```python
def bband_signal(px, N=20, K=2):
    mu = pd.rolling_mean(px, N)
    dev = pd.rolling_std(px, N)
    above = mu + K * dev
    below = mu - K * dev
    sig = DataFrame(np.where(px > above, -1., 
                             np.where(px < below, 1., 0.)),
                    px.index, px.columns)
    return sig

sig = bband_signal(px)
```

### Winsorization

```python
def winsorized(data, bounds):
    ub, lb = max(bounds), min(bounds)
    uval, lval = data.quantile(ub), data.quantile(lb)
    umask, lmask = data > uval, data < lval
    rs = data.copy()
    rs[umask] = uval
    rs[lmask] = lval
    return rs
winsorized(raw_factor, (0.005, 0.995)).hist(bins=50)
```

### Frontiers

```python
def ts_mom(px, lookback, lag):
    lagged = px.shift(lag)
    return lagged.pct_change(lookback)
```

```python
from itertools import product

def frontier(px, lb_window, lag_window):
    return {(lb, lag) : ts_mom(px, lb, lag)
            for lb, lag in product(lb_window, lag_window)}
```

```python
ser = Series(np.random.randn(1000))
sig = frontier(ser, [10, 60, 180], range(5))
sig[(10, 3)]
```
