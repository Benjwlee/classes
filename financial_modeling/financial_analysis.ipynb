### Momentum Signal Over Frontiers

```python
px = DataFrame.load('./data/stock_prices')
px = px.asfreq('B').fillna(method='ffill')
rets = px.pct_change()
```

```python
(1 + rets).cumprod().plot()
```

```python
def calc_mom(data, lookback, lag):
    mom_ret = data.shift(lag).pct_change(lookback)
    ranks = mom_ret.rank(axis=1, ascending=False)
    demeaned = ranks - ranks.mean(axis=1)
    return demeaned / demeaned.std(axis=1)
```

```python
mom = calc_mom(rets, 10, 5)
```

```python
mom.mean(axis=1).dropna() # check momentum portfolios have mean 0
```

```python
mom.std(axis=1).dropna() # check momentum portfolios have std 1
```

Run the strategy and look at the SR distribution

```python
compound = lambda x : (1 + x).prod() - 1
    
daily_sr = lambda x : x.mean() / x.std()
```

```python
def strat_sr(daily_rets, lb, hold):
    # Compute portfolio weights
    freq = '%dB' % hold
    port = calc_mom(daily_rets, lb, lag=1)
    
    #Compute portfolio returns
    port = port.shift(1).resample(freq, how='first')
    returns = daily_rets.resample(freq, how=compound)
    port_rets = (port * returns).sum(axis=1)
    
    sr = daily_sr(port_rets) * np.sqrt(252 / hold)
    return sr
```

```python
strat_sr(rets, 10, 10)
```

```python
lookbacks = range(20, 90, 5)
holdings = range(20, 90, 5)
```

```python
dd = {}
for lb in lookbacks:
    for hold in holdings:
        sr = strat_sr(rets, lb, hold)
        dd.setdefault(lb, {})[hold] = sr
```

```python
df = DataFrame(dd)
df.index.name = 'Holding Period'
df.columns.name = 'Lookback Period'
```

```python
heatmap(df)
```

### Performance Analysis

```python
# Return the portfolio weights and returns so we can do deeper analysis
def run_strat(daily_rets, lb, hold):
    freq = str(hold) + 'B'
    port = calc_mom(daily_rets, lb, lag=1)
    port = port.shift(1).resample(freq, how='first')
    returns = daily_rets.resample(freq, how=compound)
    return port, returns
```

```python
port, returns = run_strat(rets, 50, 50)
contrib = port * returns
port_rets = contrib.sum(axis=1)
```

Annualized Return

```python
port_rets.mean() * 252 / 50
```

Annualized Volatility

```python
port_rets.std() * np.sqrt(252 / 50)
```

Annualized Sharpe Ratio

```python
port_rets.mean() / port_rets.std() * np.sqrt(252 / 50)
```

Hit Rate (i.e., percent of positive periods)

```python
float((port_rets > 0).sum()) / port_rets.count()
```

```python
eu = df['EU'].dropna()
components = df[['GR', 'FR', 'IT']].dropna()
```

Use Log Returns (Siegel's Paradox)

```python
eu_rets = np.log(eu / eu.shift(1))
comp_rets = np.log(components.shift(1) / components)
```

Fixed Weights

```python
wgts = Series({'GR' : 0.5, 'FR' : 0.25, 'IT' : 0.25})
synth_rets = (comp_rets * wgts).sum(axis=1)
```
