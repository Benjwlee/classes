### Variable Weights

```python
dd = {}
for date, returns in rets.iterrows():
    port = Series(np.random.random_sample(len(rets.columns)), rets.columns)
    dd[date] = (returns * port).sum() / (port * pd.notnull(returns)).sum()
port_rets = Series(dd)
```

### Simple Stop-Loss / Take-Profit

```python
# rebalance monthly, goto 0 if performance hits +/- 3%
def make_port():
    p = Series(np.random.random_sample(len(rets.columns)), rets.columns)
    p /= p.sum()
    return p

rebal_idx = pd.period_range('1/1/1990', '7/1/2012', freq='M')
ports = DataFrame.from_dict({d: make_port() for d in rebal_idx}, orient='index')
ports = ports.reindex(rebal_idx)
ports
```

```python
mo_px = close.asfreq('M', method='ffill')
mo_rets = mo_px.pct_change(1)
mo_rets = mo_rets.to_period()
mo_rets
```

```python
def rescale(port, rets):
    ret_mask = pd.notnull(rets)
    tot_port = (ret_mask * port).sum(axis=1)
    tot_port.replace(0, nan, inplace=True)
    return tot_port

raw_port_rets = (mo_rets * ports).sum(axis=1) / rescale(ports, mo_rets)
raw_port_rets
```

```python
calc_sr(raw_port_rets, 12)
```

```python
# Naive
port_rets = raw_port_rets.copy()

stop_loss = port_rets < -0.03
take_profit = port_rets > 0.03

port_rets[stop_loss] = -0.03
port_rets[take_profit] = 0.03

calc_sr(port_rets, 12)
```

What's wrong with the naive stops?

```python
daily_ports = ports.resample('B', fill_method='ffill')

tot_port = rescale(daily_ports, rets.resample('B'))

daily_rets = (daily_ports * rets.resample('B')).sum(axis=1) / tot_port

daily_rets = daily_rets.dropna()

daily_rets
```

```python
def compound(rets):
    return (1 + rets).cumprod() - 1
```

```python
def func(daily_rets_mo, tp_level=0.03, sl_level=-0.03):
    cum_ret = compound(daily_rets_mo)
    tp = cum_ret > tp_level
    sl = cum_ret < sl_level
    if not (tp.any() or sl.any()):
        return cum_ret[-1]
    
    tp_rets = cum_ret[tp]
    sl_rets = cum_ret[sl]
    if not tp.any():
        return sl_rets[0]
    if not sl.any():
        return tp_rets[0]
    
    first_tp = tp_rets.index[0]
    first_sl = sl_rets.index[0]
    if first_tp < first_sl:
        return tp_rets[0]
    elif first_sl < first_tp:
        return sl_rets[0]
    
    raise ValueError('What is this a function for ants? We should not be here')
```

```python
strat_mo_rets = daily_rets.resample('M', how=func)
strat_mo_rets
```

```python
calc_sr(strat_mo_rets, 12)
```

```python
strat_mo_rets = daily_rets.resample('M', how=lambda x: func(x, 0.05, -0.05))

calc_sr(strat_mo_rets, 12)
```

```python
def maxdrawdown(rets):
    cum_rets = compound(rets)
    maxes = pd.rolling_max(cum_rets, len(rets), min_periods=1)
    drawdown = maxes - cum_rets
    return -drawdown[drawdown == drawdown.max()]

maxdrawdown(strat_mo_rets)
```

```python
maxdrawdown(raw_port_rets)
```

```python
def semivar(rets):
    mask = rets < 0.
    semirets = rets[mask]
    sse = (semirets**2).sum()
    
    return sse / (semirets.count() - 1)

def sortino(rets, annualizer=252.):
    return rets.mean() / np.sqrt(semivar(rets)) * np.sqrt(annualizer)

sortino(strat_mo_rets)
```
