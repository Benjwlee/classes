### Simple Linear Returns

```python
rets = px.pct_change()

compounded = (1 + rets).cumprod()
compounded.plot()
```

### Rolling window returns (With overlap)

```python
px / px.tshift(-5, freq='B') - 1. # fixed time window
```

```python
px / px.shift(5) - 1. # window within index
```

### Non-overlapping Returns

```python
mo_px = px.asfreq('MS', method='ffill')

mo_rets = mo_px.pct_change()

mo_rets
```

```python
def compound(rets):
    return (1 + rets).prod() - 1

mo_rets = px.pct_change().resample('MS', how=compound)

mo_rets
```

```python
(1 + mo_rets).cumprod().plot()
```

### Siegel's Paradox

```python
import pandas.io.data as web
usdjpy = web.get_data_fred('DEXJPUS', '1/1/1999', '6/1/2012').VALUE
usdjpy.plot()
```

```python
jpyusd = 1. / usdjpy
```

```python
df = DataFrame({'USDJPY': usdjpy.pct_change(),
                'JPYUSD': jpyusd.pct_change()})
df.head()
```

### Use log returns instead

```python
logret1 = np.log(usdjpy / usdjpy.shift(1))
logret2 = np.log(jpyusd / jpyusd.shift(1))
```

```python
df = DataFrame({'USDJPY': logret1.pct_change(),
                'JPYUSD': logret2.pct_change()})
df.head()
```
